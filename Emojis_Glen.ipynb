<b> use <i> NLTK TweetTokenizer </i></b> to normalize/tokenize texts:

```python
from nltk.tokenize import TweetTokenizer
T = TweetTokenizer(preserve_case = False, reduce_len = True, strip_handles = True)
```

```python
df.loc[12]['text']
```

```python
T.tokenize(df.loc[12]['text'])
```

Create a new column ('tokens')
Result of applying the tokenizer to each text:

```python
df['tokens'] = df['text'].apply(T.tokenize)
```

```python
subset = df['text'].str.contains('😂')
freq = pd.DataFrame({'all':Counter(concat(df['tokens'])),
                    '😂':Counter(concat(df[subset]['tokens']))})
```

```python
freq.head(10)
```

```python
freq.loc['!']
```

```python
freq.sort_values('all', ascending = False)
```

```python
freq['pmi'] = np.log2( ( freq['😂'] * np.sum(freq['all']) ) /
                       ( freq['all'] * np.sum(freq['😂']) ) )
```

```python
freq.sort_values('pmi', ascending = False)
```

```python
freq[freq['all'] > 250].sort_values('pmi', ascending = False).head(10)
```
